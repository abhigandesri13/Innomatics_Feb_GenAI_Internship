1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"


```python
#Correct Login Credentials
username = "admin"
password = "1234"
```

```python
#Asking User to Enter Details
user_name = input("Username: ")
pass_word = input("Password: ")
```

```python
#Checking the Details are Correct are not
if user_name == username:
    if pass_word == password:
        print("Login Successful")
    else:
        print("Invalid Credentials")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]


```python
#list of student marks
marks = [45, 78, 90, 33, 60]
```

```python
#Initial we take
passed = 0
failed = 0

for i in marks:
    if i >= 50:
        passed = passed + 1
    else:
        failed = failed + 1
```

```python
print("Total Students:", len(marks))
print("Total Passed:", passed)
print("Total Failed:", failed)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]



```python
# given list of names
names = [" Alice ", "bob", " CHARLIE "]
```

```python
# new list to store clean names
new_list = []

for i in names:
    i = i.strip()
    i = i.lower()
    new_list.append(i)

```

```python
#print the list
print(names)
print(new_list)
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
#list of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
```

```python
#Initially we take
error = 0

# checking each log entry
for i in logs:
    if i == "ERROR":
        error = error + 1
```

```python
#print the error
print("Total:", error)
```

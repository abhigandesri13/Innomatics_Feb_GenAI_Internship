Problem 1: Mobile Recharge Validation System

```python
# function for validation
def validate_recharge(amount, plan):

    valid_plans = [199, 299, 399, 599]

    #Recharge amount must be ≥ ₹50
    if amount <= 50:
        print("Recharge amount must be at least ₹50")
        return False

    #Valid plans: 199, 299, 399, 599
    if plan not in valid_plans:
        print("Invalid plan selected")
        return False

    print("Recharge Successful ✅")
    return True


# retry using while loop
while True:
    amt = int(input("Enter recharge amount: "))
    pln = int(input("Enter plan (199, 299, 399, 599): "))

    if validate_recharge(amt, pln):
        break
    else:
        print("Please try again \n")
```

Problem 2: Inventory Reorder Alert System

```python
# function to check stock
def check_stock(stock):
    for product, stock in stock.items():
        if stock < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK")

# products list
Products = {"Laptop": 20, "Mouse": 10, "Keyboard": 14, "Monitor": 25}

# calling the function
check_stock(Products)
```

 Problem 3: Student Result Processing System

```python
def student_result(marks):

    total = 0

    # calculate total using loop
    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# example marks list
marks_list = [60, 55, 40, 70]
result = student_result(marks_list)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
# function to calculate fare
def calculate_fare(distance, peak):

    base_fare = 50
    per_km_cost = 12

    fare = base_fare + (distance * per_km_cost)

    # adding peak hour 25% extra charge
    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


# retry system
while True:
    km = float(input("Enter distance (km): "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak_hour)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() == "no":
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
# function to check eligibility
def check_eligibility(attendance):

    present_days = 0

    # loop to count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# example attendance list
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_eligibility(attendance_list)
print(result)
```

Problem 6: Password Strength Checker

```python
# function to check password strength
def check_password(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # minimum length = 8
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # checking required characters
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# user input
pwd = input("Enter password: ")
result = check_password(pwd)

print(result)
```

Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total = 0
    for l in likes:
        total += l

    print("Total Likes:", total)

    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_check(data, required):
    total = 0
    for r in data:
        total += r

    avg = total / len(data)
    print("Average Rainfall:", avg)

    if avg >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_check([70,75,80,65], 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicates(users):
    if len(users) != len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

detect_duplicates(["abhi","raj","abhi","neha"])
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def premium_crops(prices):
    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)

premium_crops([1500,2500,1800,3200])
```

Problem 7: System Monitoring – Application Health Checker


```python
def app_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("System Status:", status)

app_health(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def exam_eligibility(attendance):
    present = 0

    for a in attendance:
        if a == "P":
            present += 1

    percent = (present / len(attendance)) * 100
    print("Attendance Percentage:", percent)

    if percent >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

exam_eligibility(["P","P","A","P","P"])
```

Problem 10: Smart Electricity Bill Analyzer


```python
def electricity_bill(units):

    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


amount, usage = electricity_bill(250)
print("Total Bill:", amount)
print("Usage Level:", usage)
```

Problem 1: Unique Words in a Sentence                                          
Problem Statement                                                   
Given a sentence, identify all unique words and display:                   
1.The count of unique words                                           
2.The unique words themselves                                            


```python
# Taking input from the user
Sentence = input("Enter Sentence: ")

#Spliting the words
words = Sentence.split()
```

```python
#Storing words in set
uniquewords = set(words)

#printing the requirements
print("Number of unique words: ", len(uniquewords))
print("Unique words: ", uniquewords)
```

Problem 2: Highest Salary from Employee Data                             
Problem Statement                                                          
Store employee names and their salaries, and determine which employee has the highest salary.                                                          


```python
# Storing the details in the dictionary
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }

# default taking values
highest_employee = ""
highest_salary = 0


```

```python
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values                                      
Problem Statement                                                               
Find the maximum and minimum values from a list of numbers.       


```python
# list of numbers
numbers = [45, 22, 89, 10, 66]

# assume first element as max and min
max_value = numbers[0]
min_value = numbers[0]
```

```python
# checking each number
for num in numbers:
    if num > max_value:
        max_value = num

    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above a Price Threshold                               
Problem Statement                                                
Count how many products have a price greater than 1000.


```python
# given prices
prices = [450, 1200, 899, 1500, 300]

# count for <1000 price
count = 0
```

```python
# checking each price
for price in prices:
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage                         
Problem Statement                                            
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
# given attendance record
attendance = ["P", "P", "A", "P", "P"]

# count present
present = 0

for day in attendance:
    if day == "P":
        present =  present + 1
```

```python
total_days = len(attendance)

attendance_percentage = (present / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

Problem 7: Count Character Frequency                                           
Problem Statement                                           
Count how many times each character appears in a given string.


```python
# given string
text = "pythonp"

# store the char
count = {}
```

```python
for char in text:
    if char in count:
        count[char] = count[char] + 1
    else:
        count[char] = 1

print(count)
```

Problem 10: Calculate Average Marks                                             
Problem Statement                                                               
Calculate the average marks of a student from a list of marks.


```python
# list of student marks
marks = [65, 70, 75, 80]

#store total marks
total = 0

for m in marks:
    total = total + m
```

```python
average = total / len(marks)

print("Average Marks:", average)
```

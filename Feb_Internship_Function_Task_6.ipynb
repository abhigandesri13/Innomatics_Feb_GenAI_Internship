Problem Statement 1:Smart Parking Lot Management System


```python
def parking_system(capacity, logs):

    parked = 0
    peak = 0

    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)


parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def delivery_time(distance, traffic, weather):

    time = distance * 5   # base 5 min per km

    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "medium":
        time += 5
    elif traffic.lower() == "low":
        time += 2

    if weather.lower() == "sunny":
        time += 2
    elif weather.lower() == "rainy":
        time += 5
    elif weather.lower() == "storm":
        time += 10

    print("Estimated Delivery Time:", time, "minutes")


delivery_time(8, "High", "Rainy")
```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def theatre_occupancy(total_seats, booked):

    occupied = len(booked)
    percent = (occupied / total_seats) * 100

    print("Occupancy:", int(percent),"%")

    if percent == 100:
        status = "Housefull"
    elif percent >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Show Status:", status)


booked_seats = [1] * 150
theatre_occupancy(200, booked_seats)
```

 Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(cpu_readings):

    total = 0
    for val in cpu_readings:
        total += val

    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)


server_load([45,60,70,85,90])
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):

    overused = []

    for name, hours in resources.items():
        if hours > 8:
            overused.append(name)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

resource_monitor(usage)
```

 Problem Statement 6: Online Event Registration Capacity Controller


```python
def event_registration(capacity, registrations):

    confirmed = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1

    waitlisted = registrations - confirmed

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if registrations >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


event_registration(100, 105)
```

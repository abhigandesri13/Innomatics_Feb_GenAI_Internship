Problem 1: Smart Light Controller


```python
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(self.name, "is", self.status)

    def turn_off(self):
        self.status = "OFF"
        print(self.name, "is", self.status)


light = SmartLight("Bedroom Light")
light.turn_on()
```

Problem 2: Employee ID Card System


```python
class EmployeeID:

    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


empolyee = EmployeeID("Rahul", "EMP102", "IT")
empolyee.display()
```

Problem 3: Mobile Contact Record


```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


contact = Contact("Anita", 9876543210)
contact.show_contact()
```

Problem 4: Product Price Tag Generator


```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def price_tag(self):
        print("Product:", self.name)
        print("Price: ₹", self.price)


p = Product("Headphones", 2499)
p.price_tag()

```

Problem 5: Movie Rating Display System


```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


movie = Movie("Inception", 4.8)
movie.display()
```

Problem 6: Delivery Address Manager


```python
class Delivery:

    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.show_details()
```
